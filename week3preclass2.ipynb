```python
# problem 4

# by using missingno
import missingno as msno
%matplotlib inline

msno.matrix(houses)
```

```python
# Check what percentage of missing values are included.

total = houses.isnull().sum().sort_values(ascending=False)
percent = (total/houses.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total,percent], axis=1,keys=['Total','percent'])
missing_data.head(20)
```

```python
# Feature quantities (columns) with five or more missing values are deleted.
houses = houses.drop(missing_data[missing_data['Total'] > 5].index,1)
# Samples (rows) with missing values are deleted.
houses = houses.drop(houses.loc[houses['Electrical'].isnull()].index)
houses.isnull().sum().max()
```

### problem 5
**skewness** is a measurement of how much a distribution leans on the right(negative leaning curve) or the left(positive leaning curve)

**kurtosis** is a measurement of the thickness of the tails of a distribution

```python
# Problem 6

import seaborn as sns

sns.distplot(houses['SalePrice'])
```

```python
print("skewness: ", houses['SalePrice'].skew())
print("kurtois: ", houses['SalePrice'].kurt())
```

```python
# Performs logarithmic transformation on the objective variable.
import numpy as np

logarithmically_transformed_houses = np.log(houses['SalePrice'])
```

```python
# The distribution is displayed for the logarithmically transformed one, and the "kurtosis" and "skewness" are also calculated.
sns.distplot(logarithmically_transformed_houses)
```

```python
print("skewness: ", logarithmically_transformed_houses.skew())
print("kurtois: ", logarithmically_transformed_houses.kurt())
```

```python
# problem 7

# Make a heat map of the correlation coefficient matrix and find the relationships with high correlation coefficients.
%matplotlib inline
import matplotlib.pyplot as plt

corrmat = houses.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True)
```

The correlation between 'GrLivArea', 'TotalBsmtSF', and 'OverallQual' is very high

```python
# Select 10 feature values that have a high correlation with the target variable, and create a correlation coefficient matrix heat map for only those 10 feature values.
number_of_variables = 10
selected_cols = corrmat.nlargest(number_of_variables,'SalePrice')['SalePrice'].index
correlation_coefficient = np.corrcoef(houses[selected_cols].values.T)
sns.heatmap(correlation_coefficient, vmax=.8, square=True, yticklabels=selected_cols.values, xticklabels=selected_cols.values,annot=True)
```
